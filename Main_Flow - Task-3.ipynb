# Data Visualization with Matplotlib

```python
#Importing the libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Reading csv file with pandas
df = pd.read_csv("C:\\Main Flow\\Main_Flow - Task-3\\householdtask3.csv")
df.head()
```

```python
df.describe()
```

*Scatter Plot*

```python
# Scatter Plot between year and income
plt.scatter(df['year'],df['income'])
# adding title
plt.title('Scatter Plot')
# adding X, Y- axis Labels
plt.xlabel('Year')
plt.ylabel('Income')
# adding more customization
plt.grid(True)
plt.colorbar()
plt.show()
```

*Line Plot*

```python
# Line Chart between income and expenditure
plt.plot(df['income'], 'b', label = 'income')
plt.plot(df['expenditure'], 'r', label = 'expenditure')
# adding title
plt.title('Line Plot')
# adding X, Y- axis Labels
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
# adding more customization
plt.grid(True)
plt.legend() 
plt.show()
```

*Bar Plot*

```python
# Bar Chart between year and income
plt.barh(df['year'], df['income'])
# adding title
plt.title('Bar Plot')
# adding X, Y- axis Labels
plt.xlabel('income')
plt.ylabel('year')
# adding more customization
plt.grid(True)
plt.show()
```

*Histogram*

```python
# Histogram
plt.hist(df['income'], bins=25, color='Brown')
# adding title
plt.title('Histogram')
# adding X, Y- axis Labels
plt.xlabel('Income')
plt.ylabel('Frequency')
# adding more customization
plt.grid(True)
plt.show()
```
